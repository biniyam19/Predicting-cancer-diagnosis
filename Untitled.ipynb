# Comparing the mean

```python
malignant = breast_cancer_data[breast_cancer_data.diagnosis == "M"].iloc[:,1:32]
benign = breast_cancer_data[breast_cancer_data.diagnosis == "B"].iloc[:,1:32]

# Malignant = breast_cancer_data[breast_cancer_data.diagnosis == "M"].iloc[:,2:]
# Benign = breast_cancer_data[breast_cancer_data.diagnosis == "B"].iloc[:,2:]
    
    
n_malignant = Malignant.shape[0]
n_benign = Benign.shape[0]
sample_size = n_malignant + n_benign 
t_test_result = []
for i in range(0, Malignant.shape[1]):
    pooled_var = ((np.var(Malignant.iloc[:,i])*(n_malignant-1)) + (np.var(Benign.iloc[:,i])*(n_benign -1))) / (sample_size -2)
    standard_error = (np.sqrt(pooled_var) * np.sqrt((1/n_malignant)) + (1/n_benign))
    mean_malignant = np.mean(Malignant.iloc[:,i])
    mean_benign = np.mean(Benign.iloc[:,i])
    mean_diff = mean_malignant - mean_benign
    lower_ci = mean_diff - 1.96*standard_error
    upper_ci = mean_diff + 1.96*standard_error
    t_test_unit = stats.ttest_ind(Malignant.iloc[:,i], Benign.iloc[:,i])
    t_test_result.append([mean_malignant, mean_benign,t_test_unit[0], t_test_unit[1], lower_ci, upper_ci])
```

```python
t_test_result = pd.DataFrame(t_test_result)
t_test_result.columns = ["Malignant","Benign","T-Value","P-Value", "Lower_CI", "Upper_CI"]
t_test_result.index=  Benign.columns
print("Table 2: Mean comparison for biomarkers by diagnosis")
print(t_test_result.to_markdown(tablefmt="grid"))
```

<div style="text-align: justify">  <p style="text-align: justify; border:3px; border-style:solid; border-color:#FF0000; padding: 1em;"> <font size="3"> T-test was used to determine if the mean biomarker values for malignant and benign tumors are statistically equal. Since the sample size is large (>30) the distribution of the t-test statistic will have approximately a normal distribution as given by the Central Limit Theorem (CLT) irrespective of the distribution of the population. The result of the t-test is presented in Table 2. At a 1% level of significance, the mean values for malignant and benign tumors are not significantly different for only four of the biomarkers: fractal_dimension_mean, fractal_dimension_se, texture_se, smoothness_se, and symmetry_se. </p>

# PCA analysis

 <div style="text-align: justify"> <font size="3"> PCA was used to further explore the relationship between biomarkers. After applying PCA, the relationship between principal components and biomarkers was analyzed to identify biomarkers that are correlated each other most and biomarkers that contributed most to the total observed variation.

```python
breast_cancer_array = breast_cancer_data_numeric.values 
min_max_scaler = preprocessing.MinMaxScaler()
breast_cancer_scaled = min_max_scaler.fit_transform(breast_cancer_array)
breast_cancer_scaled_data = pd.DataFrame(breast_cancer_scaled)
```

```python
pca = PCA(n_components=30)
pca.fit(breast_cancer_scaled_data)
pca_results = pca.transform(breast_cancer_scaled_data)
```

```python
principal_components = pd.DataFrame(pca_results)
principal_components.columns = ["PC" + str(i) for i in list(range(1,31))]
```

```python
plt.figure(figsize=(12,7))
PC_values = np.arange(pca.n_components_) + 1
plt.bar(principal_components.columns[0:10], pca.explained_variance_ratio_[0:10])
plt.title('Scree Plot', fontsize=14)
plt.xlabel('Principal Component', fontsize=14)
plt.ylabel('Proportion of Variance Explained', fontsize=14)
txt="Figure 3: Scree plot"
plt.figtext(0.5, -0.1, txt, wrap=True, horizontalalignment='center', fontsize=16)
plt.show()
```

```python
area_mean = breast_cancer_data['area_mean']
area_mean_df = pd.DataFrame(area_mean)
area_mean_df['diagnosis'] = breast_cancer_data['diagnosis']
```

<p style="text-align: justify; border:3px; border-style:solid; border-color:#FF0000; padding: 1em;"> <font size="3"> Both the scatter plot and histogram shows that there is no a single mean area value that neatly divides between malignant and benign tumors. To decide on the cutoff value various methods are available depending on what needs to be achieved and other factors.

```python
#Youden Index
```

```python
def Youden_index_calculator(x,y,n):
    """"
    prepare a dataframe to help identify cutoff based on Youden's index
    
    inputs are: 
    x = the feature to be used in finding cutoff
    y = diagnosis
    n= no of sensetivity and specificity points to be calculated   
    """
    
    #initializing 
    sen_temp = []
    spec_temp = []
    youdens_inputs = []
    cutoffs = np.linspace(start=min(area_mean), stop = max(area_mean), num = n)

    for cutoff in cutoffs:
        label_temp = area_mean.apply(lambda x: 1 if x > cutoff else 0)
        conf_mat = confusion_matrix(label_temp,diagnosis_numeric)
       
        #compute TN, FN, FP, TP
        TN = conf_mat.flatten()[0]
        FN = conf_mat.flatten()[1]
        FP = conf_mat.flatten()[2]
        TP = conf_mat.flatten()[3]
        
        #compute sensitivity and specificity
        sensitivity = TP/(TP+FN)
        specificity = TN/(TN+FP)
        
        #appending current values
        youdens_inputs.append([sensitivity, specificity, cutoff])
    
    youdens_inputs_df = pd.DataFrame(youdens_inputs)
    return youdens_inputs_df
```

```python
#Add another column -- the sum of sensitivity and specificity
youdens_inputs_df['sen_spec_sum'] = youdens_inputs_df.sensitivity + youdens_inputs_df.specificity
```

```python
# Find the index(the row), that has the maximum of the sum of the sensitivity and specificity
index_of_max = youdens_inputs_df.sen_spec_sum.idxmax()
youden_index_cutoff = youdens_inputs_df.cutoff[index_of_max]
```

```python
print("Youden index cutoff value is: {}".format(youden_index_cutoff))
```

<p style="text-align: justify; border:3px; border-style:solid; border-color:#FF0000; padding: 1em;"> <font size="3"> To find the cutoff value for the mean area Youdens index was used. Youdens index gives a cutoff value that maximizes both sensitivity and specificity. It gives a cutoff value that has the highest sensitivity + specificity. Based on the index, the cutoff value for the mean area is 696.26. Tumors with the mean area above or equal to 696.26 are malignant and those with the mean area below 696.26 are benign.
